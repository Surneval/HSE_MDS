### Input

Program asks user for data input about ticket characteristics:
1) the ticket price
2) the number of transfers
3) refund (yes or no)
4) luggage (yes or no)

Program checks wether data input is correct and transforms each input into right format.

### Main program algorythm

Program checks four parameters inputed by user and define the category of the ticket:
1) If price is less than 200; it is either a direct flight or a flight with one transfer; refund is included; luggage is included, we get **"The best offer"** category;
2) If price is in a range from 200 to 250; it's either a dicrect flight or has one or two transfers; refund is either included or not, luggage is either included or not, we get **"Good enough"** category;
3) If price is more than 250; three and more transfers; refund is either included or not, luggage is either included or not, we get **"The worst offer"**

### Output

You will get summary of the characteristics you have entered and the category of the ticket.

```python
# Ask users to input information about the ticket

ticket_price = input("Please enter the ticket price\n")
# we check whether user input is digit, if it's a digit, we transform it to int, if not - asking ti try again
if ticket_price.isdigit():
    ticket_price = int(ticket_price)
else:
    while ticket_price.isdigit() == False:
        print("Error. Type a number\n")
        ticket_price = input("Please enter the ticket price\n")
    ticket_price = int(ticket_price)

number_transfers = input("Please enter the number of transfers\n")
# we check whether user input value is digit, if it's a digit, we transform it to int, if not - asking ti try again
if number_transfers.isdigit():
    number_transfers = int(number_transfers)
else:
    while number_transfers.isdigit() == False:
        print("Error. Type a number\n")
        number_transfers = input("Please enter the number of transfers\n")
    number_transfers = int(number_transfers)

refund = input("Enter 'yes' if ticket is refundable, 'no' - if not\n")
if refund.lower() != "yes" and refund.lower() != "no":
    print("You entered incorrect values, please try again\n")
    refund = input("Enter 'yes' if ticket is refundable, 'no' - if not\n")

luggage = input("Enter 'yes' if luggage is included, 'no' - if not\n")
if luggage.lower() != "yes" and luggage.lower() != "no":
    print("You entered incorrect values, please try again\n")
    luggage = input("Enter 'yes' if luggage is included, 'no' - if not\n")

# Main program algorythm checks characteristics from input and determines the category

price1 = 200  # price bellow this limit we can consider as one of 'the best offer' factors
price2 = 250  # price over this limit we can consider as one of 'the worst offer' factors

transfer1 = 0  # number of transfers treshhold 1
transfer2 = 1  # number of transfers treshhold 2
transfer3 = 2  # number of transfers treshhold 3

best = 'the best'  # category 1
good = 'good enough'  # category 2
bad = 'bad'  # category 3
other = 'other'  # category 4

# we check all the conditions and determine the category

if ticket_price < price1 and (number_transfers == transfer1 or number_transfers == transfer2) and refund.lower() == 'yes' and luggage.lower() == 'yes':
    print(f"You have a ticket of {best} category!\nThe price is {ticket_price}.\nNumber of transfers is {number_transfers}.\nRefund and luggage are included")
elif (ticket_price >= price1 and ticket_price <= price2) and (number_transfers == transfer1 or number_transfers == transfer2 or number_transfers == transfer3) and (refund.lower() == 'yes' or refund.lower() == 'no') and (luggage.lower() == 'yes' or luggage.lower() == 'no'):
    print(f"You have a ticket of {good} category!\nThe price is {ticket_price}.\nNumber of transfers is {number_transfers}.\nRefund: {refund}. Luggage: {luggage}")
elif ticket_price > price2 and number_transfers > transfer3 and (refund.lower() == 'yes' or refund.lower() == 'no') and (luggage.lower() == 'yes' or luggage.lower() == 'no'):
    print(f"You have a ticket of {bad} category!\nThe price is {ticket_price}.\nNumber of transfers is {number_transfers}.\nRefund: {refund}. Luggage: {luggage}.")
else:
    print('You have a ticket of other category')
    
```

## Alternative solution
Let's use similar model as in the original agrerator model, we are analyzing tickets for the destination Moscow - New York.

Aggregator askes about price trashhold, noticing average price range - the first price target and the second price target, But the user still can propose his personal price range - **the first price target** and **the second price target**.

Than the agregator askes the user how many transfers are exceptable for you? User enters a number, this number is accounted in the 'best offer', number + 1 is accounted in 'good offer', bad offer has more than number + 1 transfers.

Than aggregetor askes user "Is rufund important". If the answer is "yes" - aggregator creates refund as one of the parameters of "The best" and "Good" options.

Than aggregetor askes user "Is luggage important". If the answer is "yes" - aggregator creates refund as one of the parameters of "The best" and "Good" options.

Aggregator creates the following categories:

1) "The best offer" is the one which costs less the the first price target; the number of transfers is less or equal to the number from user input, if refund is important for the user, refund is included; if luggage is important for the user the luggage is included.
2) "Good enough offer": ticket price is in range between the first price target and the second price target; the number of transfers is less or equal to the number + 2 (from user input); if luggage is important for the user thr luggage is included, refund is either included or not.
3) "Bad offer":ticket price is over the second price target; the number of transfers is more than number + 2 (from user input); luggage is either included or not, refund is either included or not.
4) "Other" if the option differs from three options above

In the output the user gets information about the category and short summary of flight details (price, number of transfers, information about the luggage and about the refund).

```python
# Alternative solution for the ticket aggregator


# Function to get data of int type from the user
def numbers_input(prompt):
    number = input(prompt)
    # we check whether user input is digit, if it's a digit, we transform it to int, if not - asking to try again
    if number.isdigit():
        number = int(number)
    else:
        while number.isdigit() == False:
            print("Error. Type a number\n")
            number = input(prompt)
        number = int(number)
    return number

# Function to get data of string type from user
def string_input(prompt):
    value = input(prompt)
    if value.lower() != "yes" and value.lower() != "no":
        while value != "yes" and value.lower() != "no":
            print("You entered incorrect values, please try again\n")
            value = input(prompt)
    return value


# Aggregator askes the user about his/her preferences and determines categories for the user

# To determine price targets program askes input from user for the first and the second price tagets
price_target1 = numbers_input(prompt="Please enter the first price target (Median is 100\n")
price_target2 = numbers_input(prompt="Please enter the second price target (more than 1st, Median is 300)\n")


# To determine transfers target program askes input from user for the acceptable number of transfers
transfers_target = numbers_input(prompt="How many transfers are exceptable for you\n")


# Program askes user about the importance of the luggage and the refund
luggage_target = string_input(prompt="Is luggage important for you?\n")
refund_target = string_input(prompt='Is refund important for you?\n')


# Program gives the summary of categoties for a certain user
best = "The best offer"
good = "Good enough offer"
bad = "Bad offer"
other = "Other"
print(f"{best} is a category with: price bellow {price_target1}, number of transfers: {transfers_target} or less (if it's more than zero), luggage: {luggage_target}, if 'no' - 'yes' option also included, refund: {refund_target}, if 'no' - 'yes' option also included")
print(f"{good} is a category with: price in range ({price_target1}, {price_target2}), number of transfers: less or equal {transfers_target + 2}, luggage: {luggage_target}, if 'no' - 'yes' option also included, refund: {refund_target}, if 'no' - 'yes' option also included")
print(f"{bad} is a category with: price above {price_target2}, number of transfers: more than {transfers_target + 2}, luggage is either included or not, refund is either included or not")
print(f"{other} - all the tickets out of three categories mentioned above")


# Program askes user to enter ticket parameters
price = numbers_input(prompt="Please enter the price of your ticket\n")
transfers = numbers_input(prompt="Please enter the number of transfers\n")
luggage_option = string_input(prompt="Enter 'yes' if luggage is included, 'no' - if not\n")
refund_option = string_input(prompt="Enter 'yes' if ticket is refundable, 'no' - if not\n")


# Program checks all the conditions and determine the category
if luggage_target.lower() == 'yes' and refund_target.lower() == 'yes':
    if price < price_target1 and transfers <= transfers_target and refund_option.lower() == refund_target.lower() and luggage_option.lower() == luggage_target.lower():
        print(f"You have a ticket of {best} category!\nThe price is {price}.\nNumber of transfers is {transfers}.\nRefund: {refund_option}. Luggage: {luggage_option}")
    elif (price >= price_target1 and price <= price_target2) and transfers <= (transfers_target +2) and refund_option.lower() == refund_target.lower() and luggage_option.lower() == luggage_target.lower():
        print(f"You have a ticket of {good} category!\nThe price is {price}.\nNumber of transfers is {transfers}.\nRefund: {refund_option}. Luggage: {luggage_option}")
    elif price > price_target2 and transfers > (transfers_target + 2) and (refund_option.lower() == 'yes' or refund_option.lower() == 'no') and (luggage_option.lower() == 'yes' or luggage_option.lower() == 'no'):
        print(f"You have a ticket of {bad} category!\nThe price is {ticket_price}.\nNumber of transfers is {number_transfers}.\nRefund: {refund}. Luggage: {luggage}.")
    else:
        print(f'You have a ticket of {other} category')
elif luggage_target.lower() == 'yes' and refund_target.lower() == 'no':
    if price < price_target1 and transfers <= transfers_target and (refund_option.lower() == refund_target.lower() or refund_option.lower() == 'yes') and luggage_option.lower() == luggage_target.lower():
        print(f"You have a ticket of {best} category!\nThe price is {price}.\nNumber of transfers is {transfers}.\nRefund: {refund_option}. Luggage: {luggage_option}")
    elif (price >= price_target1 and price <= price_target2) and transfers <= (transfers_target +2) and (refund_option.lower() == refund_target.lower() or refund_option.lower() == 'yes') and luggage_option.lower() == luggage_target.lower():
        print(f"You have a ticket of {good} category!\nThe price is {price}.\nNumber of transfers is {transfers}.\nRefund: {refund_option}. Luggage: {luggage_option}")
    elif price > price_target2 and transfers > (transfers_target + 2) and (refund_option.lower() == 'yes' or refund_option.lower() == 'no') and (luggage_option.lower() == 'yes' or luggage_option.lower() == 'no'):
        print(f"You have a ticket of {bad} category!\nThe price is {ticket_price}.\nNumber of transfers is {number_transfers}.\nRefund: {refund}. Luggage: {luggage}.")
    else:
        print(f'You have a ticket of {other} category')
elif luggage_target.lower() == 'no' and refund_target.lower() == 'yes':
    if price < price_target1 and transfers <= transfers_target and refund_option.lower() == refund_target.lower() and (luggage_option.lower() == luggage_target.lower() or luggage_option.lower() == 'yes'):
        print(f"You have a ticket of {best} category!\nThe price is {price}.\nNumber of transfers is {transfers}.\nRefund: {refund_option}. Luggage: {luggage_option}")
    elif (price >= price_target1 and price <= price_target2) and transfers <= (transfers_target +2) and refund_option.lower() == refund_target.lower() and (luggage_option.lower() == luggage_target.lower() or luggage_option.lower() == 'yes'):
        print(f"You have a ticket of {good} category!\nThe price is {price}.\nNumber of transfers is {transfers}.\nRefund: {refund_option}. Luggage: {luggage_option}")
    elif price > price_target2 and transfers > (transfers_target + 2) and (refund_option.lower() == 'yes' or refund_option.lower() == 'no') and (luggage_option.lower() == 'yes' or luggage_option.lower() == 'no'):
        print(f"You have a ticket of {bad} category!\nThe price is {ticket_price}.\nNumber of transfers is {number_transfers}.\nRefund: {refund}. Luggage: {luggage}.")
    else:
        print(f'You have a ticket of {other} category')
else:
    if price < price_target1 and transfers <= transfers_target and (refund_option.lower() == refund_target.lower() or refund_option.lower() == 'yes') and (luggage_option.lower() == luggage_target.lower() or luggage_option.lower() == 'yes'):
        print(f"You have a ticket of {best} category!\nThe price is {price}.\nNumber of transfers is {transfers}.\nRefund: {refund_option}. Luggage: {luggage_option}")
    elif (price >= price_target1 and price <= price_target2) and transfers <= (transfers_target +2) and (refund_option.lower() == refund_target.lower() or refund_option.lower() == 'yes') and (luggage_option.lower() == luggage_target.lower() or luggage_option.lower() == 'yes'):
        print(f"You have a ticket of {good} category!\nThe price is {price}.\nNumber of transfers is {transfers}.\nRefund: {refund_option}. Luggage: {luggage_option}")
    elif price > price_target2 and transfers > (transfers_target + 2) and (refund_option.lower() == 'yes' or refund_option.lower() == 'no') and (luggage_option.lower() == 'yes' or luggage_option.lower() == 'no'):
        print(f"You have a ticket of {bad} category!\nThe price is {ticket_price}.\nNumber of transfers is {number_transfers}.\nRefund: {refund}. Luggage: {luggage}.")
    else:
        print(f'You have a ticket of {other} category')
```
